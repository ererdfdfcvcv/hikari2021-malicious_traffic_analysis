# URL ANALYSIS
### Identification of malicious URL/Web-Addresses

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import plotly.express as p
import pickle

# read dataset
df = pd.read_csv("ALLFLOWMETER_HIKARI2021.csv")
df  # debugging table plotting
```

# 0. Step - Preparation of the Dataset
### - Dropping/Extraction of unnecessary features
### - Separation of the table

```python
# Drop the unnecessary features
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])

# Split the tables into 2 tables - One for the data and the other with the identifiers
df_identification = df[['uid', 'originh', 'originp', 'responh', 'responp', 'traffic_category']].copy()
df_values = df.drop(columns=['uid', 'originh', 'originp', 'responh', 'responp', 'traffic_category'])
display(df_identification)  # debugging table plotting
display(df_values)  # debugging table plotting
```

# 1. Step - Get the behavior distribution
### - Display the behavior/dataset distribution


```python
# display the distrubution of the different behavior-types
y = df_identification['traffic_category']
y.value_counts().plot.pie(autopct='%.2f')
```

```python
# display the distrubution of malicious and non-malicious entries
y = df_values['Label']
y.value_counts().plot.pie(autopct='%.2f')
```

# 2. Step - Preparation of the dataset value
### Scaling & Normalization of the data

```python
# Save the labels of the dataset
labels = df.Label
# Scaling of the data
scaler = preprocessing.StandardScaler().fit(df_values)
df_scaled = pd.DataFrame(scaler.transform(df_values))

# Normalization of the data
df_normalized_l2 = pd.DataFrame(preprocessing.normalize(df_scaled, norm='l2'))

df_normalized_l2
```

# 3. Step -  Split of dataset in Test and Train data

```python
x_train, x_test, y_train, y_test = train_test_split(df_normalized_l2, labels, test_size=0.33)
```

# 4. Setp - Build SVM Model

```python
SVM_Model = SVC(gamma='auto')
SVM_Model.fit(x_train,y_train)
```

```python
# 6. Step - Test the SVM Model
predicted_labels = SVM_Model.predict(x_test)
target_names = [
    '0',
    '1'
]
print(classification_report(y_test, predicted_labels, target_names=target_names))
with open('modelweight.pickle', 'wb') as handle:
    pickle.dump(SVM_Model, handle)
```
